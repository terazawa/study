# Chapter10: pair RDD

## Web log file

```python
logs = sc.textFile("/loudacre/weblogs/*2.log")
```

```python
logs.take(2)
```

```python
logs.map(lambda x: x.split(' ')).take(1)
```

```python
rdd = logs.map(lambda x: (x.split(' ')[2], 1))
rdd.take(1)
```

```python
rdd = rdd.reduceByKey(lambda x,y: x+y)
rdd.take(2)
```

```python
rdd1 = rdd.map(lambda t: (t[1],t[0]))
rdd1.take(2)
```

```python
rdd1.countByKey()
```

```python
rdd2 = logs.map(lambda x: x.split()).map(lambda x: (x[2],x[0]))
rdd2.take(2)
```

```python
rdd2.groupByKey().mapValues(list).collect()
```

## web log and account

```python
hivedata = sc.textFile("/user/hive/warehouse/accounts")
```

```python
rdd3 = hivedata.map(lambda x: x.split(',')).map(lambda x: (x[0],x))
rdd3.take(2)
```

```python
rdd3 = rdd3.join(rdd)
rdd3.take(2)
```
